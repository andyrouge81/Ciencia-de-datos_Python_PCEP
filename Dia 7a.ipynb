```python
import pandas as pd

# Leemos excel ficheros en su ruta
agosto2014 = pd.read_excel('/Users/andresrojo/Desktop/Python-PCEP/recursos_dia6/VentasGroucery_Meses.xlsx',
                        sheet_name= 'Ventas Agosto 2014')

septiembre2014 = pd.read_excel('/Users/andresrojo/Desktop/Python-PCEP/recursos_dia6/VentasGroucery_Meses.xlsx',
                        sheet_name= 'Ventas Septiembre 2014')

octubre2014 = pd.read_excel('/Users/andresrojo/Desktop/Python-PCEP/recursos_dia6/VentasGroucery_Meses.xlsx',
                        sheet_name= 'Ventas Octubre 2014')
octubre2014.head(5) # lo comprobamos
# unimos los dataframes en uno solo, deben coincidir en columnas, si no se crearán NAN
trimestre2014 = pd.concat([agosto2014, septiembre2014, octubre2014], ignore_index = True)
print(trimestre2014.head(5))
# formamos la columna Amount = Unit Price * Quantity

trimestre2014['Amount'] = trimestre2014['Unit Price'] * trimestre2014['Quantity']
trimestre2014
```

```python
# iloc[indice fila, indice columna]. 'iloc' es de escritura y de lectura
trimestre2014.iloc[1,1] # me devuelve el objeto contenido en la primera fila, en la primera columna
# Modificamos el interior del contenido en ese indice de la tabla
trimestre2014.iloc[1,1] = "Queso de Cabrales"
print(trimestre2014.iloc[1,1])# Visualizamos
```

```python
# Con 'iloc[]', puedo poner todas las filas y algunas de las columnas en el orden que yo quiera
# trimestre2014.iloc[:,[5,3,6,1]] en este orden indicadas con su índice
trimestre2014.iloc[:,[5,3,6,1]] # muestra todas las filas de las columnas seleccionadas
```

```python
# Recorrer todo el DataFrames, para ello utilizaremos un bucle, el bucle 'for'. Necesito saber el número de iteraciones,
# que es igual al número de filas con len()

print(len(trimestre2014), 'filas') # con len() averiguamos el número de filas

# El número de columnas 
print(len(trimestre2014.columns), 'Columnas')
```

```python
# Vamos a recorrer el Dataframe e imprimir  el producto y entre paréntesis su categoria

filas = len(trimestre2014) # número total de filas

for i in range(0,filas): # no es una 'lista' por eso hhacemos el range, filas es un número
    print(trimestre2014.iloc[i,1] + "-("+trimestre2014.iloc[i,2]+")")
```

## PRACTICA 33:
* Recorrer con un bucle (iloc) y for el dataframe trimestre2014 y cambiar el valor de la columna 'Category', para donde ponga 'Seafood' ponga _pescado_ y donde ponga 'Beverages' ponga _bebidas_.

```python
# Ponemos la longitud en 'filas' del dataset 
filas = len(trimestre2014)

for i in range(0, filas):
    if trimestre2014.iloc[i,2] == 'Seafood':# el iloc solo admite integers
        trimestre2014.iloc[i,2] = 'Pescado'
    elif trimestre2014.iloc[i,trimestre2014.columns.get_loc('Category')] == 'Beverages': # de esta forma nos retorna el índice de la columna
        trimestre2014.iloc[i,trimestre2014.columns.get_loc('Category')] = 'Bebidas'
    
trimestre2014
# Nos da el indice en funcion del nombre de la columna
print(trimestre2014.columns.get_loc('Category'))
trimestre2014
```

```python
# Las agrupaciones de un dataframe, como convertirlas a  un dataframe y como guardarlas en un fichero
# Suma del importe por categoria

trimestre2014.groupby(['Category']).Amount.sum() # suma
trimestre2014.groupby(['Category']).Amount.mean() # promedio
trimestre2014.groupby(['Category']).Amount.count() # total

# primero crear un objeto para esta agrupación
amount_category = trimestre2014.groupby(['Category']).Amount.sum()

# vamos a convertir este objeto en un dataframe de pandas
amount_category = pd.DataFrame(amount_category, columns=['Amount'])# con columns indicamos el nombre de la columna, aunque no es necesario
print(amount_category) # es un dataframe sin indice
```

```python
# Al convertir el objeto en un dataframe la columna de Category (izquierda) lo considera un 'indice'
# por lo tanto, lo que tengo que hacer es generar el indice numerico, con reset_index

amount_category = amount_category.reset_index() # coloca los indices para poder trabajar con iloc,por ejemplo
print(amount_category)
```

```python
# Guardar el dataframe en un fichero .csv 
ruta = '/Users/andresrojo/Desktop/Python-PCEP/recursos_dia7/category_group.csv'
# nombre del dataframe + .to_csv(ruta_del archivo, separacion de los valores, si queremos indices )
amount_category.to_csv(ruta, sep= ',' , index = True) 
print("Guardado con éxito")
```

```python
# Vamos a guardarlo como un .xlsx , como un excel
# nombre de la hora 'sheet_name', el nombre de las columnas, 'header 'y poner el indice con 'index'
ruta2 = '/Users/andresrojo/Desktop/Python-PCEP/recursos_dia7/category_group.xlsx'
amount_category.to_excel(ruta2, sheet_name = 'Categorias', header = True, index = True)
print('Guardado con éxito')
```

```python
# Como trabajar con 'merge'

# Conectarnos a un excel
ruta3 = '/Users/andresrojo/Desktop/Python-PCEP/recursos_dia7/Practica Merge Pandas.xlsx'
facturas = pd.read_excel(ruta3, sheet_name = 'Facturas') # seleccionamos la hoja en el fichero excel
print(facturas)

clientes =pd.read_excel(ruta3, sheet_name = 'Clientes')
print(clientes)
```

```python
# hacemos la conbinación o merge, con la función 'merge', necesitamos un objeto nuevo

datos_completos = pd.merge(facturas, clientes, on = 'IdCliente')# los unimos por el campo 'IdCliente'
print(datos_completos) # solo muestra los registros que coinciden con ambos dataframes 
# esto es un INNER JOIN (Intersección, solo muestra los registros coincidentes)
print()
datos_completos = pd.merge(facturas, clientes, on= 'IdCliente', how = 'inner')# inner es el tipo de combinación, es lo mismo que el anterior
print(datos_completos)
print()
# nombres de los campos diferentes
#datos_completos2 = pd.merge(facturas, clientes, left_on = 'IdCliente', 
                            #right_on = 'Cliente', how ='inner')# se une con las columnas de diferente nobre 'client' 'idcliente'
# si nos da error vamos a cambiar el nombre de columnas diferentes
clientes = clientes.rename(columns = {'IdClientes':'Clientes'})
datos_completos2 = pd.merge(facturas, clientes, left_on = 'IdCliente', 
                            right_on = 'Clientes', how ='inner') # da error porque no hemos añadido la columna 'Clientes' manualmente
datos_completos2
```

```python
# Ver todas las facturas, aunque no coincida con clientes, allí donde no hay registro saldrá en blanco o NAN
datos_completos_left = pd.merge(facturas, clientes, on= 'IdCliente', how = 'left')# inner es el tipo de combinación, es lo mismo que el anterior
print(datos_completos_left) # LEFT JOIN
print()
```

```python
# Ver todas las facturas, aunque no coincida con clientes, alli donde no hay registro saldrá en blanco o NAN
datos_completos_right = pd.merge(facturas, clientes, on= 'IdCliente', how = 'right')# inner es el tipo de combinación, es lo mismo que el anterior
print(datos_completos_right) # right JOIN
print()
```

```python
# Ver la combinación completa FULL OUTER JOIN , mantendrá todos los registros, alli donde no hay importe ,nombre y factura, pondrá NAN
datos_completos_outer = pd.merge(facturas, clientes, on= 'IdCliente', how = 'outer')# inner es el tipo de combinación, es lo mismo que el anterior
print(datos_completos_outer) # LEFT outer
print()
```
